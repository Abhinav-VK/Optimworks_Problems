```python
def calculate_size(file_system, path):
    # Split the path into its components
    directories = path.split('.')
    
    # Initialize the current directory to the root
    current_dir = file_system
    
    # Traverse the directory structure
    for dir_name in directories:
        if dir_name in current_dir:
            current_dir = current_dir[dir_name]
        else:
            return f"Directory '{path}' not found."
    
    # Calculate the total size of files in the current directory
    total_size = 0
    
    # Helper function to recursively calculate sizes
    def get_total_size(directory):
        nonlocal total_size
        for item, value in directory.items():
            if isinstance(value, dict):  # If it's a directory
                get_total_size(value)  # Recur into the subdirectory
            else:  # If it's a file
                total_size += value  # Add the file size

    get_total_size(current_dir)
    
    return f"Total size: {total_size}"

# Updated file system structure
filesystem = {
    "root": {
        "dir1": {
            "subdir1": {
                "file1.txt": 100,
                "file2.txt": 200,
                "subsubdir1": {
                    "file3.txt": 50,
                    "file4.txt": 150,
                    "subsubsubdir1": {
                        "file5.txt": 500,
                        "emptydir1": {}
                    }
                }
            },
            "subdir2": {
                "file6.txt": 300,
                "subsubdir2": {
                    "file7.txt": 700,
                    "subsubsubdir2": {
                        "file8.txt": 800,
                        "file9.txt": 900
                    }
                },
                "emptydir2": {}
            },
            "file10.txt": 1000
        },
        "dir2": {
            "subdir3": {
                "subsubdir3": {
                    "file11.txt": 400,
                    "file12.txt": 500
                },
                "subsubdir4": {
                    "emptydir3": {}
                }
            },
            "subdir4": {
                "file13.txt": 600,
                "subsubdir5": {
                    "file14.txt": 700,
                    "file15.txt": 800,
                    "subsubsubdir3": {
                        "emptydir4": {},
                        "file16.txt": 900,
                        "file17.txt": 1000
                    }
                }
            },
            "emptydir5": {}
        },
        "dir3": {
            "file18.txt": 1100,
            "subdir5": {
                "subsubdir6": {
                    "file19.txt": 1200,
                    "subsubsubdir4": {
                        "file20.txt": 1300,
                        "emptydir6": {}
                    }
                }
            }
        },
        "emptydir7": {},
        "file21.txt": 1400
    }
}

# Test cases
print(calculate_size(file_system, "root.dir1.subdir1"))  # Expected Output: Total size: 1000
print(calculate_size(file_system, "root.dir2.subdir4.subsubdir5"))  # Expected Output: Total size: 3400
print(calculate_size(file_system, "root.dir2.subdir4.nonexistent"))  # Expected Output: Directory 'root.dir2.subdir4.nonexistent' not found.
print(calculate_size(file_system, "root.dir1.subdir2"))  # Expected Output: Total size: 2700
```
